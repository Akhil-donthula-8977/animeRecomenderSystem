```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
anime=pd.read_csv("C:\\data science\\anime_Recomendation_dataset\\anime.csv")
ratings=pd.read_csv("C:\\data science\\anime_Recomendation_dataset\\rating.csv")
```

```python
anime=anime.dropna()
```

```python
ratings=ratings.drop(ratings[ratings["rating"]==-1].index,axis=0)
```

```python
anime
```

```python
ratings_with_name=anime.merge(ratings,on="anime_id")
```

```python
ratings_with_name
```

```python
x=ratings_with_name.groupby("user_id").count()["anime_id"].sort_values()>35
padhe_likhe_users=x[x].index
padhe_likhe_users
```

```python
## based on users
filtered_rating=ratings_with_name[ratings_with_name["user_id"].isin(padhe_likhe_users)]
```

```python
filtered_rating
```

```python
y=filtered_rating.groupby("anime_id").count()["user_id"]>50
famous_animes=y[y].index
```

```python
final_Ratings=filtered_rating[filtered_rating["anime_id"].isin(famous_animes)]
```
